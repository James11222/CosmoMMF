```python
# definitions for the axes 
left, width = 0.1, 0.65 
bottom, height = 0.2, 0.65 
spacing = 0.0

rect_center = [left, bottom, width, height]
rect_bottomx = [left, -bottom + spacing, width, 0.4] 

# start with a rectangular Figure
fig = plt.figure(figsize=(10, 6)) 
ax_center = plt.axes(rect_center) 
ax_center.tick_params(direction='in', right=True) 
ax_bottomx = plt.axes(rect_bottomx)
ax_bottomx.tick_params(direction='in') 


# the center plot:
ax_center.set_xticklabels([])
# ax_center.set_title("Isolated Components", pad = 10)
ax_center.set_ylabel("PDF")
ax_center.plot(hist_heavy_clus_TNG[1][:-1], hist_heavy_clus_TNG[0], 
               color="red", linestyle="-", label="$M_{\\rm halo} > 10^{13} M_\odot$", linewidth=3)
ax_center.plot(hist_both_clus_TNG[1][:-1], hist_both_clus_TNG[0], 
               color="coral", linestyle="-",label='$M_{\\rm halo} > 10^{12} M_\odot$', linewidth=3)
ax_center.plot(hist_all_clus_TNG[1][:-1], hist_all_clus_TNG[0], 
               color="orange", linestyle="-",label='$M_{\\rm halo} > 10^{11} M_\odot$', linewidth=3)

ax_center.plot(hist_heavy_wall_TNG[1][:-1], hist_heavy_wall_TNG[0], 
               color="green", label='Walls', linewidth=3, linestyle=":")

ax_center.plot(hist_heavy_fil_TNG[1][:-1], hist_heavy_fil_TNG[0], 
               color="blue", label='Heavy Filaments', linewidth=3, linestyle=linestyles["loosely dashed"])
ax_center.plot(hist_both_fil_TNG[1][:-1], hist_both_fil_TNG[0], 
               color="royalblue", label='Light Filaments', linewidth=3, linestyle="--")
ax_center.plot(hist_all_fil_TNG[1][:-1], hist_all_fil_TNG[0], 
               color="deepskyblue", label='Ultralight Filaments', linewidth=3, linestyle=linestyles["densely dashed"])



ax_center.plot(hist_heavy_void_TNG[1][:-1], hist_heavy_void_TNG[0], 
               color="purple", label='Voids', linewidth=3, linestyle="-.")


ax_center.set_ylim(1.2e-3, 5e3)
ax_center.set_xlim(6e-3, 1.4e2)
ax_center.grid()
ax_center.legend(shadow=True, edgecolor="black", ncol = 2, fontsize=15)
ax_center.semilogy()
ax_center.semilogx()

#redidual plot

ax_bottomx.set_ylabel("$\mathrm{PDF}_{\mathrm{hydro}} \; / \; \mathrm{PDF}_{\mathrm{dm}} $")
ax_bottomx.set_xlabel("$\\rho/ \\bar{\\rho}$")

ax_bottomx.plot(hist_heavy_clus_dm[1][:-1], hist_heavy_clus_TNG[0] / hist_heavy_clus_dm[0],
                color="red", linestyle="-", linewidth=3)
ax_bottomx.plot(hist_both_clus_dm[1][:-1], hist_both_clus_TNG[0] / hist_both_clus_dm[0],
                color="coral",linestyle="-", linewidth=3)
ax_bottomx.plot(hist_all_clus_dm[1][:-1],hist_all_clus_TNG[0] / hist_all_clus_dm[0],
                color="orange",linestyle="-", linewidth=3)

ax_bottomx.plot(hist_heavy_fil_dm[1][:-1], hist_heavy_fil_TNG[0] / hist_heavy_fil_dm[0], color="blue",
                linestyle=linestyles["loosely dashed"], linewidth=3)
ax_bottomx.plot(hist_both_fil_dm[1][:-1], hist_both_fil_TNG[0] / hist_both_fil_dm[0], color="royalblue",
                linestyle="--", linewidth=3)
ax_bottomx.plot(hist_all_fil_dm[1][:-1], hist_all_fil_TNG[0] / hist_all_fil_dm[0], color="deepskyblue",
                linestyle=linestyles["densely dashed"], linewidth=3)

ax_bottomx.plot(hist_heavy_wall_dm[1][:-1], hist_heavy_wall_TNG[0] / hist_heavy_wall_dm[0], color="green",
                linestyle=":", linewidth=3)
ax_bottomx.plot(hist_heavy_void_dm[1][:-1], hist_heavy_void_TNG[0] / hist_heavy_void_dm[0], color="purple",
                linestyle="-.", linewidth=3)



# ax_bottomx.hlines(y = 1, xmin = rng[0], xmax=rng[1], color="gray", linestyle="--", linewidth=3)
# ax_bottomx.set_xlim(rng[0], rng[1])
ax_bottomx.set_xlim(6e-3, 1.4e2)
ax_bottomx.set_ylim(-0.1, 1.4) 
# ax_bottomx.semilogy()
# ax_center.set_xticklabels(labelspad=10)
ax_bottomx.semilogx()
ax_bottomx.grid()


# fig.subplots_adjust(bottom=-10) # or whatever
fig.subplots_adjust(bottom=-100)

# fig.savefig("Figures/Final_PDF.pdf",bbox_inches="tight", dpi=300)
# plt.tight_layout()

```
