## Isolated Component Analysis

```python
den_TNG = np.load("/global/cscratch1/sd/james12/simple_density_cubes/density_cube_snap_099_full_1024_all_species.npy")

μ_den_TNG = np.mean(den_TNG)
den_TNG = (den_TNG - μ_den_TNG) / μ_den_TNG
```

```python
heavy_den_TNG_clus = den_TNG[heavy_clusbool_TNG]
print("Made heavy Clusters")
del heavy_clusbool_TNG

light_den_TNG_clus = den_TNG[light_clusbool_TNG]
print("Made light Clusters")
del light_clusbool_TNG

ultralight_den_TNG_clus = den_TNG[ultralight_clusbool_TNG]
print("Made ultra light Clusters")
del ultralight_clusbool_TNG

heavy_den_TNG_fil = den_TNG[heavy_filbool_TNG]
print("Made heavy Filaments")
del heavy_filbool_TNG

light_den_TNG_fil = den_TNG[light_filbool_TNG]
print("Made light Filaments")
del light_filbool_TNG

ultralight_den_TNG_fil = den_TNG[ultralight_filbool_TNG]
print("Made ultralight Filaments")
del ultralight_filbool_TNG

heavy_den_TNG_wall = den_TNG[heavy_wallbool_TNG]
print("Made Walls")
del heavy_wallbool_TNG

heavy_den_TNG_void = den_TNG[heavy_voidbool_TNG]
print("Made Voids")
del heavy_voidbool_TNG

del den_TNG

```

```python
den_dm = np.load("/global/cscratch1/sd/james12/simple_density_cubes/density_cube_snap_099_full_1024_dm.npy")

μ_den_dm = np.mean(den_dm)
den_dm = (den_dm - μ_den_dm) / μ_den_dm
```

```python
heavy_clusbool_dm = np.load("/global/cscratch1/sd/james12/NEXUS_analysis/DM_only_1024_snap_099/heavy_halos/density_cube_snap_099_full_1024_dm_cluster_bool_filter.npy")
light_clusbool_dm = np.load("/global/cscratch1/sd/james12/NEXUS_analysis/DM_only_1024_snap_099/light_halos/density_cube_snap_099_full_1024_dm_cluster_bool_filter.npy")
ultralight_clusbool_dm = np.load("/global/cscratch1/sd/james12/NEXUS_analysis/DM_only_1024_snap_099/ultralight_halos/density_cube_snap_099_full_1024_dm_cluster_bool_filter.npy")
heavy_filbool_dm = np.load("/global/cscratch1/sd/james12/NEXUS_analysis/DM_only_1024_snap_099/heavy_halos/density_cube_snap_099_full_1024_dm_filament_bool_filter.npy")
light_filbool_dm = np.load("/global/cscratch1/sd/james12/NEXUS_analysis/DM_only_1024_snap_099/light_halos/density_cube_snap_099_full_1024_dm_filament_bool_filter.npy")
ultralight_filbool_dm = np.load("/global/cscratch1/sd/james12/NEXUS_analysis/DM_only_1024_snap_099/ultralight_halos/density_cube_snap_099_full_1024_dm_filament_bool_filter.npy")

heavy_wallbool_dm = np.load("/global/cscratch1/sd/james12/NEXUS_analysis/DM_only_1024_snap_099/heavy_halos/density_cube_snap_099_full_1024_dm_wall_bool_filter.npy")

heavy_voidbool_dm = ~(heavy_clusbool_dm + heavy_filbool_dm + heavy_wallbool_dm)
```

```python
heavy_den_dm_clus = den_dm[heavy_clusbool_dm]
print("Made Heavy Clusters")
del heavy_clusbool_dm

light_den_dm_clus = den_dm[light_clusbool_dm]
print("Made Light Clusters")
del light_clusbool_dm

ultralight_den_dm_clus = den_dm[ultralight_clusbool_dm]
print("Made Ultra Light Clusters")
del ultralight_clusbool_dm

heavy_den_dm_fil = den_dm[heavy_filbool_dm]
print("Made Filaments")
del heavy_filbool_dm

light_den_dm_fil = den_dm[light_filbool_dm]
print("Made light Filaments")
del light_filbool_dm

ultralight_den_dm_fil = den_dm[ultralight_filbool_dm]
print("Made ultralight Filaments")
del ultralight_filbool_dm

heavy_den_dm_wall = den_dm[heavy_wallbool_dm]
print("Made Walls")
del heavy_wallbool_dm

heavy_den_dm_void = den_dm[heavy_voidbool_dm]
print("Made Voids")
del heavy_voidbool_dm

del den_dm
```

```python
rng = (-1.5,10)

hist_heavy_clus_TNG = np.histogram((heavy_den_TNG_clus.flatten()), bins = 100, density=True, range=rng)
hist_light_clus_TNG = np.histogram((light_den_TNG_clus.flatten()), bins = 100, density=True, range=rng)
hist_ultralight_clus_TNG = np.histogram((ultralight_den_TNG_clus.flatten()), bins = 100, density=True, range=rng)

hist_heavy_fil_TNG = np.histogram((heavy_den_TNG_fil.flatten()), bins = 100, density=True, range=rng)
hist_light_fil_TNG = np.histogram((light_den_TNG_fil.flatten()), bins = 100, density=True, range=rng)
hist_ultralight_fil_TNG = np.histogram((ultralight_den_TNG_fil.flatten()), bins = 100, density=True, range=rng)


hist_heavy_wall_TNG = np.histogram((heavy_den_TNG_wall.flatten()), bins = 100, density=True, range=rng)
hist_heavy_void_TNG = np.histogram((heavy_den_TNG_void.flatten()), bins = 100, density=True, range=rng)
```

```python
hist_heavy_clus_dm = np.histogram((heavy_den_dm_clus.flatten()), bins = 100, density=True, range=rng)
hist_light_clus_dm = np.histogram((light_den_dm_clus.flatten()), bins = 100, density=True, range=rng)
hist_ultralight_clus_dm = np.histogram((ultralight_den_dm_clus.flatten()), bins = 100, density=True, range=rng)

hist_heavy_fil_dm = np.histogram((heavy_den_dm_fil.flatten()), bins = 100, density=True, range=rng)
hist_light_fil_dm = np.histogram((light_den_dm_fil.flatten()), bins = 100, density=True, range=rng)
hist_ultralight_fil_dm = np.histogram((ultralight_den_dm_fil.flatten()), bins = 100, density=True, range=rng)


hist_heavy_wall_dm = np.histogram((heavy_den_dm_wall.flatten()), bins = 100, density=True, range=rng)
hist_heavy_void_dm = np.histogram((heavy_den_dm_void.flatten()), bins = 100, density=True, range=rng)
```

```python
# definitions for the axes 
left, width = 0.1, 0.65 
bottom, height = 0.2, 0.65 
spacing = 0.0

rect_center = [left, bottom, width, height]
rect_bottomx = [left, -bottom + spacing, width, 0.4] 

# start with a rectangular Figure
fig = plt.figure(figsize=(10, 6)) 
ax_center = plt.axes(rect_center) 
ax_center.tick_params(direction='in', right=True) 
ax_bottomx = plt.axes(rect_bottomx)
ax_bottomx.tick_params(direction='in') 


# the center plot:
ax_center.set_xticklabels([])
ax_center.set_title("Isolated Components", pad = 10)
ax_center.set_ylabel("PDF")
ax_center.plot(hist_heavy_clus_TNG[1][:-1], hist_heavy_clus_TNG[0], 
               color="red", label="$M_{\\rm halo} > 10^{13} M_\odot$", linewidth=3)
ax_center.plot(hist_light_clus_TNG[1][:-1], hist_light_clus_TNG[0], 
               color="coral", label='$M_{\\rm halo} = [10^{12} M_\odot, 10^{13} M_\odot]$', linewidth=3)
ax_center.plot(hist_ultralight_clus_TNG[1][:-1], hist_ultralight_clus_TNG[0], 
               color="orange", label='$M_{\\rm halo} = [10^{11} M_\odot, 10^{12} M_\odot]$', linewidth=3)

ax_center.plot(hist_heavy_wall_TNG[1][:-1], hist_heavy_wall_TNG[0], 
               color="green", label='Walls', linewidth=3, linestyle=":")

ax_center.plot(hist_heavy_fil_TNG[1][:-1], hist_heavy_fil_TNG[0], 
               color="blue", label='Heavy Filaments', linewidth=3, linestyle="--")
ax_center.plot(hist_heavy_fil_TNG[1][:-1], hist_heavy_fil_TNG[0], 
               color="royalblue", label='Light Filaments', linewidth=3, linestyle="--")
ax_center.plot(hist_heavy_fil_TNG[1][:-1], hist_heavy_fil_TNG[0], 
               color="deepskyblue", label='Ultralight Filaments', linewidth=3, linestyle="--")



ax_center.plot(hist_heavy_void_TNG[1][:-1], hist_heavy_void_TNG[0], 
               color="purple", label='Voids', linewidth=3, linestyle="-.")


ax_center.set_xlim(rng[0], rng[1])
ax_center.grid()
ax_center.legend(shadow=True, edgecolor="black", ncol = 2, fontsize=15)


#redidual plot

ax_bottomx.set_ylabel("$\mathrm{PDF}_{\mathrm{hydro}} \; / \; \mathrm{PDF}_{\mathrm{dm}} $")
ax_bottomx.set_xlabel("$(\\rho - \\bar{\\rho}) / \\bar{\\rho}$")

ax_bottomx.plot(hist_heavy_clus_dm[1][:-1], hist_heavy_clus_TNG[0] / hist_heavy_clus_dm[0],
                color="red", linewidth=3)
ax_bottomx.plot(hist_light_clus_dm[1][:-1], hist_light_clus_TNG[0] / hist_light_clus_dm[0],
                color="coral", linewidth=3)
ax_bottomx.plot(hist_ultralight_clus_dm[1][:-1],hist_ultralight_clus_TNG[0] / hist_ultralight_clus_dm[0],
                color="orange", linewidth=3)

ax_bottomx.plot(hist_heavy_fil_dm[1][:-1], hist_heavy_fil_TNG[0] / hist_heavy_fil_dm[0], color="blue",
                linestyle="--", linewidth=3)
ax_bottomx.plot(hist_light_fil_dm[1][:-1], hist_light_fil_TNG[0] / hist_light_fil_dm[0], color="royalblue",
                linestyle="--", linewidth=3)
ax_bottomx.plot(hist_ultralight_fil_dm[1][:-1], hist_ultralight_fil_TNG[0] / hist_ultralight_fil_dm[0], color="deepskyblue",
                linestyle="--", linewidth=3)

ax_bottomx.plot(hist_heavy_wall_dm[1][:-1], hist_heavy_wall_TNG[0] / hist_heavy_wall_dm[0], color="green",
                linestyle=":", linewidth=3)
ax_bottomx.plot(hist_heavy_void_dm[1][:-1], hist_heavy_void_TNG[0] / hist_heavy_void_dm[0], color="purple",
                linestyle="-.", linewidth=3)



ax_bottomx.hlines(y = 1, xmin = rng[0], xmax=rng[1], color="gray", linestyle="--", linewidth=3)
ax_bottomx.set_xlim(rng[0], rng[1])
ax_bottomx.set_ylim(0, 2.4)
ax_bottomx.grid()

# fig.subplots_adjust(bottom=-10) # or whatever
fig.subplots_adjust(bottom=-100)

# fig.savefig("Figures/PDF_isolated.png",bbox_inches="tight", dpi=300)
# plt.constrain_layout()

```
