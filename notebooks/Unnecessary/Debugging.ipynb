```julia
wave_vecs = CosmoMMF.wavevectors3D((100,100,100))

log_smooth_sphere_field = CosmoMMF.smooth_loggauss(sphere_field, 0.5, wave_vecs)
log_smooth_cylinder_field = CosmoMMF.smooth_loggauss(cylinder_field, 0.5, wave_vecs)
log_smooth_wall_field = CosmoMMF.smooth_loggauss(wall_field, 0.5, wave_vecs)
print(np.min(log_smooth_sphere_field))

#Generating the very large figure
#one figure with 2 columns, and 3 rows

f = plt.figure(figsize=(10,16))
f.suptitle("Density Fields - log-smoothed", fontsize=30)
#top left
ax1 = plt.subplot(321)
ax1.imshow(log_smooth_sphere_field[:,50,:])
ax1.set_title("Sphere-xz")
ax1.set_xlabel("x-axis")
ax1.set_ylabel("z-axis")
im1 = ax1.imshow(log_smooth_sphere_field[:,50,:])
divider1 = axgrid.make_axes_locatable(ax1)
cax1 = divider1[:append_axes]("right", size="5%", pad=0.05)
plt.colorbar(im1, cax=cax1)
#top right
ax2 = plt.subplot(322)
ax2.imshow(log_smooth_sphere_field[:,:,50])
ax2.set_title("Sphere-xy")
ax2.set_xlabel("x-axis")
ax2.set_ylabel("y-axis")
#mid left
ax3 = plt.subplot(323)
ax3.imshow(log_smooth_cylinder_field[:,50,:])
ax3.set_title("Cylinder-xz")
ax3.set_xlabel("x-axis")
ax3.set_ylabel("z-axis")
#mid right
ax4 = plt.subplot(324)
ax4.imshow(log_smooth_cylinder_field[:,:,50])
ax4.set_title("Cylinder-xy")
ax4.set_xlabel("x-axis")
ax4.set_ylabel("y-axis")
#bottom left
ax5 = plt.subplot(325)
ax5.imshow(log_smooth_wall_field[:,50,:])
ax5.set_title("Wall-xz")
ax5.set_xlabel("x-axis")
ax5.set_ylabel("z-axis")
#bottom right
ax6 = plt.subplot(326)
ax6.imshow(log_smooth_wall_field[:,:,50])
ax6.set_title("Wall-xy")
ax6.set_xlabel("x-axis")
ax6.set_ylabel("y-axis")

f.tight_layout()
#plt.savefig("Density_Fields_log_smoothed.png", format="png", dpi=300, bbox_inches="tight")
plt.show()
```
