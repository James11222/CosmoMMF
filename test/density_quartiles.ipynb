```python
import numpy as np
from pylab import *
from scipy.ndimage import gaussian_filter
%matplotlib inline
```

```python
test_img = np.random.normal(size=(16,16,16))
test_img = gaussian_filter(test_img, sigma=3)
```

```python
# imshow(test_img, interpolation='nearest', cmap='inferno')
# colorbar()
```

```python
q1, q2, q3 = np.quantile(test_img, [0.25, 0.5, 0.75])
print (q1, q2, q3)
```

```python
structure_class = np.zeros(shape=test_img.shape)
structure_class [np.where (test_img<q1)] = 0
structure_class [np.where ((test_img<q2) & (test_img>q1))] = 1
structure_class [np.where ((test_img<q3) & (test_img>q2))] = 2
structure_class [np.where (test_img>q3)] = 3
```

```python
figure(figsize=(12,4))
subplot(121)
imshow(test_img[:,:,1], interpolation='nearest')
title('input density field')
colorbar()
subplot(122)
imshow(structure_class[:,:,1], interpolation='nearest', cmap='inferno')
title('structure classified by its density quartiles')
colorbar()
```
